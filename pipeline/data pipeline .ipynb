```python
file_name = "fromgithub"
file_path = f"Files/data/{file_name}"
df = spark.read.csv(file_path,header = True,inferSchema = True)
df.show()
```

```python
display(df)
```

```python
from pyspark.sql.functions import col,when
df_transform = df.withColumn(
    "subject1_status",
    when(col("subject_1") >= 35,"pass").otherwise("Fail")
    ).withColumn(
        "subject2_status",
        when(col("subject_2") >= 35,"pass").otherwise ("Fail")
        ).withColumn(
            "subject3_status",
            when(col("subject_3") >= 35,"pass").otherwise("Fail")
            )
df_transform.show()
```

```python
display(df_transform)
```

```python
from delta.tables import DeltaTable
df_transform.write.option("overwriteSchema","true").format("delta").mode("overwrite").save("Tables/students_data")
```
